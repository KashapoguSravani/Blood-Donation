```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import json, pandas as pd

# Replace with your actual file path
path = "snips_dataset/2017-06-custom-intent-engines/AddToPlaylist/train_AddToPlaylist.json"

with open(path, 'r') as f:
    data = json.load(f)

# Handle dictionary or list structures automatically
if isinstance(data, dict):
    for key, value in data.items():
        if isinstance(value, list):
            df = pd.DataFrame(value)
            print(f" Extracted {len(df)} samples from key '{key}'")
            break
    else:
        df = pd.DataFrame([data])
elif isinstance(data, list):
    df = pd.DataFrame(data)
else:
    raise ValueError(" Unsupported JSON structure")

print("Columns:", df.columns.tolist())
df.head()
```

```
import matplotlib.pyplot as plt

if 'intent_name' in df.columns:
    df['intent_name'].value_counts().plot(kind='bar', title='Intent Distribution', figsize=(10,4))
    plt.xlabel("Intent Name")
    plt.ylabel("Count")
    plt.show()
else:
    print(" 'intent_name' column not found  here are your columns:", df.columns.tolist())
```

```
print("Total samples:", len(df))
print("Unique intents:", df['intent_name'].nunique() if 'intent_name' in df else "N/A")
print("Intent names:", df['intent_name'].unique() if 'intent_name' in df else "N/A")
```

```
def combine_text(tokens):
    """Combine token dictionaries into a single sentence."""
    if isinstance(tokens, list):
        return "".join([t.get('text', '') for t in tokens])
    return ""

# Apply the function to rebuild sentences
df['text'] = df['data'].apply(combine_text)

# Assign intent name manually (since this file is from AddToPlaylist)
df['intent_name'] = 'AddToPlaylist'

# Drop the old nested column
df = df[['text', 'intent_name']]

print(" Cleaned DataFrame:")
print(df.head(5))
print(f"\nTotal samples: {len(df)}")
```

```
import os, json, pandas as pd

all_records = []

base_dir = "snips_dataset/2017-06-custom-intent-engines/"
for intent_dir in os.listdir(base_dir):
    path = os.path.join(base_dir, intent_dir, f"train_{intent_dir}.json")
    if not os.path.exists(path):
        continue

    with open(path, 'r') as f:
        data = json.load(f)

    if isinstance(data, dict):
        for key, value in data.items():
            if isinstance(value, list):
                temp_df = pd.DataFrame(value)
                temp_df['text'] = temp_df['data'].apply(lambda x: "".join([t.get('text', '') for t in x]))
                temp_df['intent_name'] = key
                all_records.append(temp_df[['text', 'intent_name']])
                break

# Combine all intents
df = pd.concat(all_records, ignore_index=True)
print(" Combined all intents. Shape:", df.shape)
df.head()
```

```
import matplotlib.pyplot as plt

df['intent_name'].value_counts().plot(kind='bar', figsize=(10,4), title='Intent Distribution')
plt.xlabel("Intent Name")
plt.ylabel("Sample Count")
plt.show()
```
